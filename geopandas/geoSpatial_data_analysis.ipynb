## filter and plot specific municipality

```python
aathrai_tribeni = gdf[gdf['GaPa_NaPa']=='Aathrai Tribeni']
```

## calculate area and peimeter of each municipality

```python
gdf['area_sqkm'] = gdf.area / 1e6
gdf['Perimeter_km'] = gdf.length /1000

gdf[['GaPa_NaPa', 'area_sqkm', 'Perimeter_km']]
```
